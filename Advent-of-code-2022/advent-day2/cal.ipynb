```python
with open('input.txt') as f:
    lines = [line.rstrip('\n') for line in f]
```

```python
res = 0

for i in lines:
    round = 0
    opp, me = i[0], i[-1]

    #Add my chose
    if me == "X": round += 1
    elif me == "Y": round += 2
    elif me == "Z": round += 3


    #Add win darw lose
    if me == "X" and opp == "A": round += 3
    elif me == "X" and opp == "B": round += 0
    elif me == "X" and opp == "C": round += 6

    if me == "Y" and opp == "B": round += 3
    elif me == "Y" and opp == "C": round += 0
    elif me == "Y" and opp == "A": round += 6

    if me == "Z" and opp == "C": round += 3
    elif me == "Z" and opp == "A": round += 0
    elif me == "Z" and opp == "B": round += 6

    res += round
   
```

```python
# part 2 
totP2 = 0
for i in lines:
    round = 0
    opp, traget = i[0], i[-1]

    if traget == "X": 
        # lose
        if opp == "A": round += 3
        elif opp == "B": round += 1
        elif opp == "C": round += 2

    elif traget == "Y":
        #Draw 
        if opp == "A": round += 1
        elif opp == "B": round += 2
        elif opp == "C": round += 3
        round += 3
    
    elif traget == "Z": 
        # win
        if opp == "A": round += 2
        elif opp == "B": round += 3
        elif opp == "C": round += 1
        round += 6
   
    totP2 += round
    print(round)
```
