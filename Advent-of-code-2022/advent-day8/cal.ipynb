```python
import pandas as pd
import csv
import numpy as np
```

```python
df = pd.read_csv("p_input.csv", header=None)
arr = df.to_numpy()
```

```python
def is_vis_left(row, col, arr):
    t = arr[row,col]
    if col == 0:
        return True
    
    l = max(arr[row,:col])
    print(l,t)
    return l < t
    
```

```python
def is_vis_rigth(row, col, arr):
    num_rows, num_cols = arr.shape
    t = arr[row,col]
    if col == num_cols-1:
        return True
    
    r = max(arr[row, col+1:])
    print(r,t)
    return r < t
```

```python
def is_vis_up(row, col, arr):
    t = arr[row,col]
    if row == 0:
        return True
    
    up = max(arr[:row, col])
    print(up,t)
    return up < t
```

```python
def is_vis_down(row, col, arr):
    t = arr[row,col]
    num_rows, num_cols = arr.shape
    if row == num_rows-1:
        return True
    down = max(arr[row+1:, col])
    print(down,t)
    return down < t
```

```python
num_rows, num_cols = arr.shape
res = 0
count = 0

for r in range(num_rows):
    for c in range(num_cols):
        if is_vis_down(r,c,arr) or is_vis_up(r,c,arr) or is_vis_left(r,c,arr) or is_vis_rigth(r,c,arr):
            res += 1
        count+=1
```
